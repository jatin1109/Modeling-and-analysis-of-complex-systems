Our model simulates the commuter traffic in a single circular bus route. To keep track of our model, we need the number of people waiting at each station and the number of people inside each of the bus. This information is the state of the system. We have a rather unreasonable assumption that the number of people who arrive at any station is independent of the time of the day. We also assume that the arrival distribution at each station is independent and identical to distribution at other stations. One more assumption is that the passengers only travel a maximum of 7 stops. Another reasonable assumption is that while people disembark, nobody can embark on the bus. Our assumption about the capacity of a bus being 130 is a bit unrealistic, but we'll roll with it. We also make more assumptions such as the stop at wich a passenger gets off, the distribution of time it takes to disembark and embark, and the travel time between stops. We also make an assumption that the passengers will be completely rational and choose to board a bus even if there is only 1 seat left. There is nobody who hates crowded spaces. After each unit of time, there will be some new arrivals at the stops. They will add to the number of people at the stops. Whereas, when a person boards a bus, they leave the stop. Similarly, whenever the bus reaches a stop, first, a number of passangers get off and some number of passengers get on. This changes the number of people in the bus. The people who exit the bus have now exited the system. Our inputs will be all the distribution parameters for the normal distributions, exponential distributions. We would also need the number of buses, the number of stops and the state of the system initially. Our output will be the state of the system at a desired time.

```python
#Exercise 2

import numpy as np
import scipy.stats as sts
rate_parameter = 3               #This is known as lambda in the exponential distribution   
distribution = sts.expon(scale= 1/rate_parameter)
sample = distribution.rvs(size = 1000)
#part 1
exp_mean = np.mean(sample)
exp_var = np.var(sample)
mean_actual = 1/rate_parameter
var_actual = 1/(rate_parameter**2)

print("Sample mean = ", exp_mean, "\n Actual mean = ", mean_actual)
print("Sample variance = ", exp_var, "\n Actual variance = ", var_actual)
```

Part 2:


Yes, the sample mean and variance match the theoretical values for this distribution. Although they are not exactly the same, they are close enough. This is what we should expect since we are sampling from the distribution itself and our sample statistics should near the theoretical statistics as our sample size increases.
