```python
import random
import numpy as np
import scipy.stats as sts
import math
import matplotlib.pyplot as plt
```

```python
def flip(p):
    move = 0
    if random.random() < p:
        move = 1
    else:
        move = -1
    return move
number_of_step = 30
flips = []
position = 0
for i in range(number_of_step):
    position = position + flip(0.6)
    flips.append(position)
print("The final position is", flips[-1])
plt.plot(list(range(number_of_step)),flips)
plt.xlabel("Steps")
plt.ylabel("Distance from starting point")
```

```python
N = 200
final_position_list = []
for j in range(N):
    position = 0
    for i in range(number_of_step):
        position = position + flip(0.6)
    final_position_list.append(position)
print(final_position_list)
plt.hist(final_position_list)
plt.xlabel("Distance")
plt.ylabel("Frequency")
```

```python
print("The sample mean is", np.mean(final_position_list))
```

```python
print("The variance is", np.var(final_position_list))
```

```python
def gambler_ruin(limit, gambler, house):
    gambler_fortunes = [gambler]
    while gambler_fortunes[-1] > 0 and gambler_fortunes[-1] < (gambler + house):
        if limit and len(gambler_fortunes) == limit:
            break
        w = random.choice([-1, 1])
        gambler_fortunes.append(gambler_fortunes[-1] + w)
    return gambler_fortunes
```

```python
limits = [10000, 100000, 1000000, None]
for lim in limits:
  bankruptcy = []
  for i in range(1000):
      arr = gambler_ruin(lim, 100, 200)
      if arr[-1] == 0:
          bankruptcy.append(len(arr))
  print(f"For limit = {lim}, the mean is {np.mean(bankruptcy)}, and the variance is {np.var(bankruptcy)}")
```
