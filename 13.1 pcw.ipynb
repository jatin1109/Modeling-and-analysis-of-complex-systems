If a network has parameters n, q, ⟨k⟩, then, the number of nodes in the LCC is given by (1-q)×n. The number of edges will be approximately ⟨k⟩×(1-q)×n

```python
import matplotlib.pyplot as plt
import numpy as np

def calculate_q(k):
    '''
    Use a numerical root finder to determine q from the equation
    q = exp(k*(q­1)).
    '''
    from scipy.optimize import root
    return root(lambda q: q - np.exp(k * (q - 1)), 0).x[0]

def lcc_size(n, k):
    q = calculate_q(k)
    return (1 - q) * n
```

```python
n = 100
k_val = np.linspace(1, 10, 100)
lcc_sizes = [lcc_size(n, k) for k in k_val]

plt.figure()
plt.plot(k_val, lcc_sizes)
plt.xlabel("Degree of network")
plt.ylabel("Number of Nodes in LCC network")
plt.show()
```
