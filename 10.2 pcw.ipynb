```python
import tkinter
import matplotlib
matplotlib.use('TkAgg')
from pylab import *
import networkx as nx
import random as rd
```

```python
def initialize():
    global g
    g = nx.karate_club_graph()
    g.pos = nx.spring_layout(g)
    for i in g.nodes:
        g.nodes[i]['state'] = 1 if random() < .5 else 0

def observe():
    global g
    cla()
    nx.draw(g, vmin = 0, vmax = 1,
            node_color = [g.nodes[i]['state'] for i in g.nodes],
            pos = g.pos)

def update():
    global g
    listener = rd.choice(list(g.nodes))
    speaker = rd.choice(list(g.neighbors(listener)))
    g.nodes[listener]['state'] = g.nodes[speaker]['state']

def check_converge():
    global g
    states = [] 
    for node in g.nodes():
        states.append(node['state'])
        
    if sum(states) == len(g.nodes()):
        return True 
    
    elif sum(states) == 0:
        return True 
    
    else:
        return False


convergence_numbers = []
n = 1
for i in range(n):
    initialize()       
    steps = 0       
    while check_converge() == False:
        update()            
        steps += 1
        if steps % 1000 == 0:
            print(f"Reached {steps} steps")
    convergence_numbers.append(steps)

plt.figure()
plt.hist(convergence_numbers, density = True)
plt.title("Exercise 16.2")
plt.xlabel("Convergence steps")
print(f"Average time length is {np.mean(convergence_numbers)}")
```

```python
def reversed_update():
    global g
    speaker = rd.choice(list(g.nodes))
    listener = rd.choice(list(g.neighbors(speaker)))
    g.nodes[listener]['state'] = g.nodes[speaker]['state']

convergence_numbers = []
n = 200
for i in range(n):
    initialize()       
    steps = 0       
    while check_converge() == False:
        reversed_update()            
        steps += 1
    convergence_numbers.append(steps)

plt.figure()
plt.hist(convergence_numbers, density = True)
plt.title("Exercise 16.3 reversed update")
plt.xlabel("Convergence steps")
print(f"Average time length is {np.mean(convergence_numbers)}")
```
